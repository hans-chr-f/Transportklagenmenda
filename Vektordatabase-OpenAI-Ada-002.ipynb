```python
import json
import time

BATCH_SIZE = 100

# Load data from a JSON file
json_file_path = 'vedtak.json'  # replace with your JSON file path
with open(json_file_path, 'r', encoding='utf-8') as file:
    file_data = json.load(file)

documents = []
metadatas = []
ids = []

collection = client.get_collection(name="Vedtak_hele_ada_002", embedding_function=openai_ef)

for index, data in enumerate(file_data):
    documents.append(data['Vedtak_komplett'])
    # Append additional metadata here
    metadata = {
        'saksnummer': data['Saksnummer'],
        'dato': data.get('Dato', ''),
        'tjenesteytere': data.get('Tjenesteytere', ''),
        'nemnd': data.get('Nemnd', ''),
        'har_mindretall': data.get('Har_mindretall', '')
    }
    metadatas.append(metadata)

    # Debugging: Print metadata for every entry
    #print(f"Metadata for entry {index + 1}: {metadata}")

    ids.append(str(index + 1))

    if (index + 1) % BATCH_SIZE == 0 or index == len(file_data) - 1:
        # Debugging: Print metadata before adding to collection
        #print(f"Adding the following metadata to collection: {metadatas}")

        collection.add(
            documents=documents,
            metadatas=metadatas,
            ids=ids
        )
        print(f"Added {index + 1} documents out of {len(file_data)}")
        time.sleep(10)
        # Clear the lists after adding to collection
        documents.clear()
        metadatas.clear()
        ids.clear()
```

```python
import numpy as np


# Load Chromadb collection
collection = client.get_collection("Vedtak_hele_ada_002", embedding_function=openai_ef)

# Get the embeddings for the support documents from the collection
support_embeddings = collection.get(include=['embeddings'])['embeddings']

dists = collection.query(query_embeddings=support_embeddings, n_results=2, include=['distances'])

# Flatten the distances list, excluding the first element (which is an element's distance to itself)
flat_dists = [item for sublist in dists['distances'] for item in sublist[1:]]

from umap.umap_ import UMAP
import umap.plot as umap_plot
import numpy as np

mapper = UMAP().fit(support_embeddings)
umap_plot.points(mapper, values=np.array(flat_dists), show_legend=False, theme='inferno')
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

collection = client.get_collection("Vedtak_hele_ada_002", embedding_function=openai_ef)


# Get the embeddings for the support documents from the collection
support_embeddings = collection.get(include=['embeddings'])['embeddings']

# Get the 'metadatas' for each point
metadata_list = collection.get(include=['metadatas'])['metadatas']

# Extract 'nemnd' values from the metadata list
nemnd_values = [meta['nemnd'] for meta in metadata_list]

# Define an integer mapping for each nemnd value
nemnd_keywords = ["Fly", "Sjø", "Kollektivreiser", "Pakkereisenemnda"]
int_map = {"Fly": 0, "Sjø": 1, "Kollektivreiser": 2, "Pakkereisenemnda": 3}
int_values = np.array([int_map[val] for val in nemnd_values])

# Define a custom colormap
colors = ["red", "green", "blue", "purple"]
custom_cmap = ListedColormap(colors)

from umap.umap_ import UMAP

mapper = UMAP().fit(support_embeddings)

# Create the UMAP plot
fig, ax = plt.subplots(figsize=(10, 10))  # Adjust figure size
scatter = ax.scatter(
    mapper.embedding_[:, 0], 
    mapper.embedding_[:, 1], 
    c=int_values, 
    cmap=custom_cmap,
    s=1,           # Point size
    alpha=0.8      # Transparency
)

# Create a legend for the custom colormap
handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=col, markersize=10) for col in colors]
ax.legend(handles, nemnd_keywords, title="Nemnd Keywords")

plt.show()


```
